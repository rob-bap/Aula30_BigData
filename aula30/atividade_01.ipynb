```python

import pandas as pd
import numpy as np

ENDERECO_DADOS = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'
```

```python
try:
    df_roubos_celular = pd.read_csv(ENDERECO_DADOS, sep= ';', encoding='iso-8859-1')
    # display(df_roubos)
    # display(df_roubos.columns)
    # display(df_roubos['regiao'].unique())
    # display(df_roubos.head())

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
try:
    df_roubos_celular = df_roubos_celular[['cisp', 'regiao', 'munic', 'roubo_celular']]

    df_correcao_acentuacao = df_roubos_celular['regiao'].str.startswith('Grande Niter', na=False)
    # display(df_correcao_acentuacao)

    # Grava Grande Niterói na posição onde for encontrada
    df_roubos_celular.loc[df_correcao_acentuacao, 'regiao'] = 'Grande Niterói'
    # display(df_roubos['regiao'].unique())

    # Agrupando por cisp, regiao e munic
    df_roubo = df_roubos_celular.groupby(['cisp', 'regiao', 'munic'], as_index= False)['roubo_celular'].sum()

    display(df_roubo)

except Exception as e:
    print(f'Erro ao limpar os dados: {e}')
```

```python
try:

    df_roubos_celular = np.array(df_roubo['roubo_celular'])

    media = np.mean(df_roubos_celular)
    mediana = np.median(df_roubos_celular)
    total = np.sum(df_roubos_celular)

    # obtendo os quartis
    q1 = np.quantile(df_roubos_celular, 0.25)
    q3 = np.quantile(df_roubos_celular, 0.75)

    print('##### MEDIDAS ######')
    print(f'Média: {media}')
    print(f'Mediana: {mediana}')
    print(f'Total: {total}')
    print(f'Q1: {q1}')
    print(f'Q3: {q3}')

except Exception as e:
    print(f'Erro ao processar as informações: {e}')
```

```python
try:

    # Maiores
    df_maiores = df_roubo[df_roubo['roubo_celular'] > q3].copy()
    df_maiores['flag'] = 'Maiores'

    # Menores
    df_menores = df_roubo[df_roubo['roubo_celular'] < q1].copy()
    df_menores['flag'] = 'Menores'

    # Concatenando os DataFrames
    df_roubos_flag = pd.concat([df_maiores, df_menores], ignore_index=True)
    display(df_roubos_flag)

except Exception as e:
    print(f'Erro ao criar a coluna flag: {e}')
```

```python
try:

    df_roubos_flag.to_csv('roubos_veiculos.csv', index=False)
    df_roubos_flag.to_excel('roubos_veiculos.xlsx', index=False)
    print('DataFrame exportado com sucesso!')


except Exception as e:
    print(f'Erro ao exportar dados: {e}')
```
